```python
import json

# Opening JSON file
f = open('annotations.json')

# returns JSON object as 
# a dictionary
data = json.load(f)
  
# Iterating through the json
# list
text = data['text']
annotation_tuples = []

# Convert annotations to Spacy format
for annotation in data['highlights']:
    begin = annotation['comment']['begin']
    end = annotation['comment']['end']
    tag = annotation['comment']['text']
    annotation_tuples.append((begin, end, tag))

# Spacy training data
training_data = [
    (text, annotation_tuples)
]

print(training_data)
  
# Closing file
f.close()
```
